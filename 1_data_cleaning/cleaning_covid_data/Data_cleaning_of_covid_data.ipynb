```python
# importing libraries
import pandas as pd

# reading the data
data=pd.read_csv("owid-covid-data.csv")

# extracting only relevant columns
df=data[["new_cases","date","location"]]
# converting date to datetime
df["date"] =  pd.to_datetime(df["date"])

# extracting only dates relevant to our study
df=df[(df['date'] >= '2020-03-28') & (df['date'] <= '2020-05-01')]

# only getting the global data
df=df[(df['location'] == 'World')]

# calculating the incidence rate, which is new cases per 10000
df['new_cases']=df['new_cases'].div(10000)

# renaming the column to reflect the incidence rate measurement
df=df.rename(columns={'new_cases': 'covid_cases_per_10000'})

# Saving file to csv
df.to_csv("COVID_cases.csv")
df
```
